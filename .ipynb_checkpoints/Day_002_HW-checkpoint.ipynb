## 作業
請找出一個你認為機器學習可以發光發熱的領域並闡述原因?

Ans. 社會工作領域

李開復博士曾說，目前機器學習較能看出成效的地方，是在小的場域內，蒐集大量的資料，產生可供決策的預測模型

我目前的工作領域是在非營利組織，也就是未來人工智慧取代不了的有愛的服務業

無庸置疑的，社會工作是直接人與人之間面對面的服務，看似機器學習在這塊領域上派不上用場

但其實不然，目前社會工作除了經驗法則之外，也漸漸建構出了一套方法論

社工人員除了情感上的服務，也必須紀錄每次去案家訪視所蒐集到的資訊，進而擬定下一次需著重的處遇及目標

若能夠用機器學習的，依照多年來累積的數據資料，可產生預測模型，供社工人員做為評估參考用

比如說，依據家庭相關的背景、人口組成、收入支出，可預測出這個家庭是否需要經濟上的幫助

透過分析每次的工作紀錄，可以預測出可能是高風險家庭的機率是多少

其實社會工作，仍有很多可以量化與評估的準則，若能夠利用機器學習來幫助社工人員處理量化的評估

可以讓社工人員更著重人工智慧無法達到的，專精在人與人之間情感上的服務
