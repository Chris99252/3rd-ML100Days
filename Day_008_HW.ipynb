```python
# Import 需要的套件
import os
import numpy as np
import pandas as pd

# 設定 data_path
dir_data = './data/'
```

```python
f_app_train = os.path.join(dir_data, 'application_train.csv')
app_train = pd.read_csv(f_app_train)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
print('平均值 = ', app_train['AMT_INCOME_TOTAL'].mean())
print('標準差 = ', app_train['AMT_INCOME_TOTAL'].std())
print('---- 四分衛數 ----')
print(app_train['AMT_GOODS_PRICE'].quantile(np.arange(0,1.25,0.25)))
```

```python
floatlis = []
for i in app_train.columns:
    if app_train[i].dtype in ['float64']:
        floatlis.append(i)
```

```python
plt.figure(figsize=(20,20))
for i in range(len(floatlis)):
    plt.subplot(9, 8, i + 1)
    plt.hist(x = app_train[floatlis[i]])
    plt.title(floatlis[i], size = 8)
```
